```python
from neural_style_fn import neural_style
from glob import glob
import numpy as np
```

```python
content_dir = 'C:/Users/guest-developer/Google Drive/Research/Metastable Impressions/Data/Visualisations/target_samples/'
content_imgs = glob(content_dir+'*')
```

```python
style_dir = 'C:/Users/guest-developer/Google Drive/Research/Metastable Impressions/Background/timeline_images/'
style_imgs = glob(style_dir+'*')
```

```python
output_dir ='C:/Users/guest-developer/Google Drive/Research/Metastable Impressions/Data/Visualisations/output/'
```

```python
all_inputs = []
for content_img in content_imgs:
    for style_img in style_imgs:
        output = output_dir+content_img.split('\\')[-1].split('.')[0]+'-'+style_img.split('\\')[-1].split('.')[0]+'.jpg'
        if not 'desktop' in output:
            all_inputs.append((content_img,style_img,output))
```
